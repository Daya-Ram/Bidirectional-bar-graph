```python
import matplotlib.pyplot as plt
import numpy as np

# Data
y_labels = ['0-4','5-9','10-14','15-19','20-24','25-29','30-34','35-39','40-44','45-49','50-54','55-59','60-64','65-69','70-74','75-79','80-84','85-89','90-94','95']
y_Male = [1278760.4,1430246.73,1482316.69,1480898.73,1289157.75,1112011.49,970051.52,928389.81,820940.34,681257.43,686181.13,532657.55,461714.22,376227.70,289391.59,169439.16,77601.07,36551.72,13070.75,6747.92]
y_Female = [1138285.77,1310961.16,1401021.60,1458463.14,1468531.51,1324917.75,1158081.65,1094491.67,910958.18,741691.43,714782.00,533478.00,485178.36,388356.12,314423.30,180543.99,82482.16,41061.23,17058.06,10328.97]

# Create figure and axes with a larger size
fig, ax = plt.subplots(figsize=(20, 16))  # Adjust the size as needed

# Define the bar width
bar_width = 0.6

# Define the x locations for the groups
index = np.arange(len(y_labels))

# Plot Male data
bars1 = ax.barh(index, y_Male, bar_width, label='Male', color='blue')

# Plot Female data with negative values
bars2 = ax.barh(index, [-x for x in y_Female], bar_width, label='Female', color='pink')

# Set labels, title, and legend
ax.set_xlabel('Population of the year 2020', fontsize=16, color="blue")
ax.set_ylabel('Age Group', fontsize=16, color="green")
ax.set_title('Population by Age Group and Gender', fontsize=14, color="Maroon")
ax.set_yticks(index)
ax.set_yticklabels(y_labels, color="purple")  # Set y-tick color

# Set custom x-labels
custom_x_labels = ['0', '1', '2', '3', '4', '5', '6']
custom_x_positions = np.linspace(0, max(max(y_Male), max(y_Female)), len(custom_x_labels))
ax.set_xticks(custom_x_positions)
ax.set_xticklabels(custom_x_labels, color="blue")

# Invert x-axis to start from the right
ax.invert_xaxis()

# Adding values as text in a vertical line for Male
for i, bar in enumerate(bars1):
    ax.text(bar.get_width(), bar.get_y() + bar.get_height() / 2, f'{y_Male[i]:,.2f}', 
            va='center', ha='right',fontsize=14, color="green")

# Adding values as text in a vertical line for Female
for i, bar in enumerate(bars2):
    ax.text(bar.get_width(), bar.get_y() + bar.get_height() / 2, f'{y_Female[i]:,.2f}', 
            va='center', ha='left',fontsize=14, color="Maroon")

# Adjust legend position
plt.legend(loc='upper left', bbox_to_anchor=(0.90, 0.95))

# Set the text "Index" inside the graph
ax.text(0.95, 0.96, "Index", ha="right", va="top", transform=ax.transAxes, fontsize=14, color="blue", weight="bold")

# Set x-tick color
ax.tick_params(axis='x', colors='blue')

# Show plot
plt.show()
```
